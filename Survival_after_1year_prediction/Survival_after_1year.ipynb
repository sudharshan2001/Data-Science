```python
import pandas as pd
import seaborn as sns
import matplotlib as plt
from sklearn.metrics import accuracy_score
import numpy as np

df = pd.read_csv('Training_set_intermediate.csv')
pd.set_option('display.max_columns',None)
```

### You can find the dataset here
https://www.kaggle.com/rsrishav/patient-survival-after-one-year-of-treatment 

```python
dataf=df.copy()
dataf.head(5)
```

### Finding Unique values

```python
for i in dataf.columns:
    print(f'The unique values in {i} are: ', dataf[i].unique(), '----' ,' and the length is' ,len(dataf[i].unique()),'\n')
```

```python
dataf.info()
```

```python
dataf['Treated_with_drugs'].value_counts()
#DX6 is the most treated drugs 
```

```python
#As DX6 is the most treated drugs we're filling null values with them
dataf['Treated_with_drugs'].fillna('DX6',inplace=True)
dataf['Treated_with_drugs'].isnull().any()
```

# Finding mode values and replacing the NaN with it

```python
# Checking the mode of the columns which has null values
for i in dataf.iloc[:,9:16].columns:
    print(f"{i}'s value counts are\n" , dataf[i].value_counts())
    print(f"{i}'s mode is", dataf[i].mode().item())
    print('________________')
```

```python
# Filling the Null values with mode values
for i in dataf.iloc[:,9:16].columns:
    dataf[i].fillna(dataf[i].mode().item(),inplace=True)
```

```python
# making sure they're filled
for i in dataf.iloc[:,9:16].columns:
    print(f"{i}'s value counts are\n" , dataf[i].value_counts())
    print(f"{i}'s mode is", dataf[i].mode().item())
    print('________________')
```

## Dropping unwanted features

```python
# Dropping "Number of previous condition" rows which has null values 
dataf.dropna(how='any',subset=['Number_of_prev_cond'],inplace=True)
```

```python
dataf.head(1)
```

```python
# we don't neen mental condition as all of it are stable
dataf.drop(['ID_Patient_Care_Situation','Patient_ID','Patient_mental_condition'],inplace=True,axis=1)
```

```python
#  As majority of the patients in this dataset don't smoke we're filling that in null value places
dataf['Patient_Smoker'].replace({'Cannot say': 'NO'},inplace=True)
```

```python
# Removing this as there are only 3 patients treated with this
dataf=dataf[dataf['Treated_with_drugs']!='DX1 DX2 DX3 DX4 DX5 ']
```

# Making a balanced dataset

```python
data_surv1 = dataf[dataf['Survived_1_year']==1]
data_surv0 = dataf[dataf['Survived_1_year']==0]
```

```python
data_surv1.shape,data_surv0.shape
```

```python
data_surv1 = data_surv1.head(8207)
data_surv1.shape,data_surv0.shape
```

```python
# concatenating them
dataf=pd.concat([data_surv1, data_surv0], ignore_index=True)
dataf.tail(1)
```

# Making OHE on categorical features

```python
encoded_dataf = pd.get_dummies(data=dataf, columns=['Treated_with_drugs', 'Patient_Smoker','Patient_Rural_Urban'])
```

```python
encoded_dataf.head(2)
```

## Arranging the order of features

```python
encoded_columns = encoded_dataf.columns.tolist()[:11]+encoded_dataf.columns.tolist()[12:]+encoded_dataf.columns.tolist()[11:12]
encoded_dataf=encoded_dataf[encoded_columns]
```

```python
# making the dtype as int32
encoded_dataf[['A','B','C','D','E','Z','Number_of_prev_cond','Patient_Body_Mass_Index']]=encoded_dataf[['A','B','C','D','E','Z','Number_of_prev_cond','Patient_Body_Mass_Index']].astype('int32')
```

```python
encoded_dataf.head(1)
```

# Training our model

### classifying features and labels

```python
for_x = encoded_columns[0:-1]
x = encoded_dataf[for_x]
for_y = encoded_columns[46:]
y = encoded_dataf[for_y]
```

```python
# splitting the data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.15, random_state=7)
```

#  Linear Regression

```python
from sklearn.linear_model import LogisticRegression
model_lore = LogisticRegression(max_iter=150)
```

```python
model_lore.fit(x_train,y_train)
```

```python
y_predicted = model_lore.predict(x_test)
```

```python
score = accuracy_score(y_test, y_predicted)
score
```

# Decision Tree

```python
from sklearn import tree
model_tree = tree.DecisionTreeClassifier()
```

```python
model_tree.fit(x_train,y_train)
```

```python
y_predicted_tree = model_tree.predict(x_test)
accuracy_score(y_test, y_predicted_tree)
```

# Neural Nets 

```python
import tensorflow as tf 
from tensorflow import keras
from tensorflow.keras import layers
from sklearn.preprocessing import normalize
from tensorflow.keras.models import Sequential 
from tensorflow.keras.layers import Dense,Activation,Dropout 
```

```python
inputs = keras.Input(shape=(46))
x = layers.Dense(512, activation="relu", name="first_layer")(inputs)
x = layers.Dropout(0.2)(x)
x = layers.Dense(256, activation="relu", name="second_layer")(x)
x = layers.Dense(128, activation="relu")(x)

outputs = layers.Dense(2, activation="softmax")(x)
model = keras.Model(inputs=inputs, outputs=outputs)

model.compile(
    loss=keras.losses.SparseCategoricalCrossentropy(from_logits=False),
    optimizer=keras.optimizers.Adam(lr=0.001),
    metrics=["accuracy"],
)
```

```python
model.summary()
```

```python
model.fit(x_train,y_train , batch_size=32, epochs=20, validation_data=(x_test,y_test),verbose=2)
```
